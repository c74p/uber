# Part 2

### Question
Part 2 ‚Äê Experiment and metrics design
The neighboring cities of Gotham and Metropolis have complementary circadian rhythms: on
weekdays, Ultimate Gotham is most active at night, and Ultimate Metropolis is most active
during the day. On weekends, there is reasonable activity in both cities.
However, a toll bridge, with a two way toll, between the two cities causes driver partners to tend
to be exclusive to each city. The Ultimate managers of city operations for the two cities have
proposed an experiment to encourage driver partners to be available in both cities, by
reimbursing all toll costs.
1. What would you choose as the key measure of success of this experiment in
encouraging driver partners to serve both cities, and why would you choose this metric?
2. Describe a practical experiment you would design to compare the effectiveness of the
proposed change in relation to the key measure of success. Please provide details on:
a. how you will implement the experiment
b. what statistical test(s) you will conduct to verify the significance of the
observation
c. how you would interpret the results and provide recommendations to the city
operations team along with any caveats.

### Response

1. Unfortunately, the question doesn't include the financials involved. In particular, how much is a ride worth to us, and
how much does a toll cost?  There are lots of variables that have (surely intentionally) been left out.

The simplest metric would measure independence of the events that a ride is accepted by the closest driver, and that a 
rider is on the other side of the bridge from the driver.  In other words,

**P(closest driver accepts the request | rider is on the opposite side) / P(closest driver accepts the request)**

where "closest" is defined by either distance or travel time, whichever metric is already in use. Total success with this metric is 1 and total failure is 0. (This implicitly assumes that we're at 0 today, even though that's not exactly true.)

As simple and clean as that metric is, drivers are humans and not always rational; so it's reasonably likely that you'll
have some drivers that change, and some that don't. There may be a need to monitor a group of drivers to see how they
respond, or maybe even a desire to interview some drivers; for any of these reasons, focusing the metric on the driver may 
be more useful:

**P(driver accepts the closest request | rider is on the opposite side) / P(driver accepts the closest request)

With this metric, you could arrive at the aggregate measure above, and do lots of slicing and dicing; plus it comes with
an individual goal that could even go into a driver's reporting app. And, you could give it a cheeseball name like the
"tollmaster score." So this is probably the best metric. And for purposes of this discussion, I'll pretend like
"tollmaster score" is a good name even though it's not, and use it.

2. A simple experiment would be to identify a group of drivers today with same or similar tollmaster scores and do an A/B
test.
a. Of course, the treatment and control groups are chosen at random from the group of drivers to take part in the test. 
The control group gets no special attention. The treatment group gets the toll reimbursement, along with an internal
marketing campaign, which frankly should probably even include some initial incentives to the driver beyond just "free 
tolls". In any case, they get the full-nine-yards rollout, hopefully including some gamification in their app about who 
gets to be the "tollmaster" (I'm thinking the old Ghostbusters theme here, but there's probably something more culturally
relevant).

b. We take the mean of the "tollmaster scores" and do a test-of-means analysis. We should get 100 or so drivers total, so we'd have 50 drivers in each group and could just do a standard test of means.

c. If the difference in means was significant at 95% confidence level (p < 0.05), I'd feel comfortable saying it worked
for the moment.

I would certainly take a look at the metric before, during, and after the experiment, for different reasons:
* Before is just to have a baseline.
* During is to get a handle on whether/how there was a build in the changes.  Did the rollout work?  Was it too short or
too long?  Did it have a strong initial impact and then tail off, or did it have a slow start and take a while to take
hold?  If the latter, was there anything specific that changed, or was it a cumulative effect?
* After is because many drivers, even once they've changed their habits, may change back. There's an effect when they know
you're watching - it's very common in the telematics world.  So, it's important to continue monitoring to see whether the
results continue.  And, if they start to tail off, what then?  Are we as an organization ready to support a re-
implementation campaign?  How much is it worth to the business?
